```python
df_train.Text = df_train.Gene +" "+ df_train.Variation +" "+ df_train.Text
```

```python
#Define preprocessing function

import nltk
from nltk.corpus import stopwords
newstopwords=stopwords.words("English")
WNlemma = nltk.WordNetLemmatizer()

def pre_process(text):
    tokens = nltk.word_tokenize(text)
    tokens=[WNlemma.lemmatize(t) for t in tokens]
    tokens=[word for word in tokens if word not in newstopwords]
    text_after_process=" ".join(tokens)
    return(text_after_process)
```

```python
#Apply the function on each document
df_train['Text'] = df_train['Text'].apply(pre_process)
```

```python
#split the data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_train[['Gene', 'Variation', 'Text']], df_train.Class, test_size=0.2, random_state=12)
```

<H1>Preprocessing</H1>

```python
#Create dtm by using word occurence

from sklearn.feature_extraction.text import CountVectorizer


count_vect = CountVectorizer( )

X_train_counts = count_vect.fit_transform(X_train.Text)
X_train_counts.shape
```

```python
count_vect.get_feature_names()
```

```python
#Create dtm by using Term Frequency. 
#Divide the number of occurrences of each word in a document 
#by the total number of words in the document: 
#these new features are called tf for Term Frequencies
#If set use_idf=True, which mean create dtm by using tf_idf

from sklearn.feature_extraction.text import TfidfTransformer
tf_transformer = TfidfTransformer(use_idf=True).fit(X_train_counts)
X_train_tf = tf_transformer.transform(X_train_counts)
X_train_tf.shape
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
le = LabelEncoder()
enc = OneHotEncoder(handle_unknown='ignore')
gene_encoded = le.fit_transform(X_train.Gene)
gene_encoded = gene_encoded.reshape(len(gene_encoded), 1)
X_train_gene = enc.fit_transform(gene_encoded)
X_train_gene.shape
```

```python
le_v = LabelEncoder()
enc_v = OneHotEncoder(handle_unknown='ignore')
var_encoded = le_v.fit_transform(X_train.Variation)
var_encoded = var_encoded.reshape(len(var_encoded), 1)
X_train_var = enc_v.fit_transform(var_encoded)
X_train_var.shape
```

```python
from scipy.sparse import hstack
X_train = hstack((X_train_gene,X_train_var,X_train_tf))
```

```python
X_train
```

```python
from sklearn.preprocessing import LabelBinarizer

lbl = LabelBinarizer(sparse_output=True)
y_train = lbl.fit_transform(y_train)
y_train.shape
```

```python
#Building Modeling by using Naïve Bayes
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(X_train_tf, y_train)
```

```python
X_test_counts = count_vect.transform(X_test.Text)
X_test_tf = tf_transformer.transform(X_test_counts)

y_test = lbl.transform(y_test)
```

```python
X_test_counts = count_vect.transform(X_test.Text)
X_test_tf = tf_transformer.transform(X_test_counts)

gene_encoded_test = le.transform(X_test.Gene)
gene_encoded_test = gene_encoded_test.reshape(len(gene_encoded_test), 1)
X_test_gene = enc.transform(gene_encoded_test)

var_encoded_test = le_v.transform(X_test.Variation)
var_encoded_test = var_encoded_test.reshape(len(var_encoded_test), 1)
X_test_var = enc_v.fit_transform(var_encoded_test)

X_test = hstack((X_test_gene,X_test_var,X_test_tf))

y_test = lbl.transform(y_test)
```

```python
#Test model accuracy
from sklearn import metrics 
predicted = clf.predict(X_test_tf)
print(metrics.confusion_matrix(lbl.inverse_transform(y_test), predicted))
print(np.mean(predicted == lbl.inverse_transform(y_test)) )
#print(metrics.log_loss(lbl.inverse_transform(y_test), predicted))
```
